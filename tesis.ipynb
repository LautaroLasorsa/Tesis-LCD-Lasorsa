## La varianza como variable

```python
datos_varianzas = pd.read_csv("datos_varianzas.csv", sep='\t')
```

```python
datos_varianzas
```

```python
interes = [x  for x in range(10000,1_000_001) if 1000000%x==0]

def Test(datos_varianza, varianza):
    datos_loc = datos_varianza[datos_varianza['-1']==varianza]
    return [np.corrcoef(datos_loc['1'], datos_loc[str(tamGrupo)])[0][1] for tamGrupo in interes]
    
```

```python
observaciones = list(zip(*[Test(datos_varianzas, var) for var in datos_varianzas['-1'].unique()]))
```

```python
observaciones
```

```python
plt.figure(figsize = (20,12))
plt.plot([1000000//x for x in interes], list(observaciones), '-o')
plt.xlabel("Número de grupos")
plt.ylabel("Correlacion lineal con valor idealmente medido")
plt.title("Poblacion lognormal de 1.000.000 individuos")
plt.xticks(rotation = 90)
plt.legend(datos_varianzas['-1'].unique(), title = "Varianza de la \nNormal asociada", )
plt.grid(True)
plt.show()
```
